```python
# importing modules
import re
import json
import requests
from datetime import datetime
import csv

# define URL To begin scraping
url = 'https://shopee.com.my/Mega-Deal-M-5XL-Men-Short-Pants-Sport-Shorts-Beach-Shorts-Casual-Fashion-Men-Pants-Seluar-Pendek-Lelaki-Men\'s-Pant-i.138500883.12855869293?sp_atk=cfa43c5f-3567-49d4-84b8-97e64dfa7804&xptdk=cfa43c5f-3567-49d4-84b8-97e64dfa7804'

# allows for extraction of ID's from the URL
r = re.search(r'i\.(\d+)\.(\d+)', url)
if r:
    shop_id, item_id = r.group(1), r.group(2)
else:
    print("Invalid URL")
    exit()

# Create a csv file with headers
headers = ['Username', 'Date', 'Comment']
with open('ratings.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(headers)

# Offset essentially allows us to keep track of what we have scraped in this instance ratings and while true means infinite loop until end of time or until there is no more rates
# for rating we extract username date and comment
offset = 0
while True:
    ratings_url = f'https://shopee.com.my/api/v2/item/get_ratings?filter=0&flag=1&itemid={item_id}&limit=20&offset={offset}&shopid={shop_id}&type=0'
    data = requests.get(ratings_url).json()

    for rating in data['data']['ratings']:
        username = rating['author_username']
        date = datetime.fromtimestamp(rating['ctime'])
        comment = rating['comment']

        # Write the data to the csv file 
        with open('ratings.csv', 'a', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerow([username, date, comment])

        print(username)
        print("DATE: [", date, "]")
        print(comment)
        print('-' * 80)

# if the number of rating is less than 20 loop breaks
    if len(data['data']['ratings']) < 20:
        break

    offset += 20
```
